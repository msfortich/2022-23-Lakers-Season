```python
import pandas as pd
import json
```

<h1>Cleaning Game Info and Converting JSON to CSV<h1>

```python
raw_data = open("../raw/lakergames2022.json")
raw_games = json.load(raw_data)
games = []
for game in raw_games['data']:
        game_data = {
            'Date': game['date'],
            'Home Team': game['home_team']['full_name'],
            'Home Score': game['home_team_score'],
            'Visitor Team': game['visitor_team']['full_name'],
            'Visitor Score': game['visitor_team_score'],
        }
        games.append(game_data)
laker_games = pd.DataFrame(games)
laker_games['Date'] = pd.to_datetime(laker_games['Date']).dt.strftime('%Y/%m/%d')
lakers_2022 = laker_games.loc[0:82].sort_values(by = 'Date')
lakers_2022 = lakers_2022.drop(80) # Playoff Game
print(lakers_2022.isnull().info())
lakers_2022
```

```python
lakers_2022.to_csv('clean_lakers_games.csv')
```

<h1>Cleaning Game Stats and Converting to new CSV<h1>

```python
raw_stats = pd.read_csv("../raw/lakers_2022_stats.csv")
raw_stats['GAME_DATE'] = pd.to_datetime(raw_stats['GAME_DATE']).dt.strftime('%Y/%m/%d')
raw_stats.columns
```

```python
stats = raw_stats[['SEASON_YEAR','TEAM_ABBREVIATION', 'TEAM_NAME',
       'GAME_DATE', 'MATCHUP', 'WL', 'MIN', 'FGM', 'FGA', 'FG_PCT', 'FG3M',
       'FG3A', 'FG3_PCT', 'FTM', 'FTA', 'FT_PCT', 'OREB', 'DREB', 'REB', 'AST',
       'TOV', 'STL', 'BLK', 'BLKA', 'PF', 'PFD', 'PTS', 'PLUS_MINUS']]
stats = stats.rename(columns={'SEASON_YEAR': 'Season','TEAM_ABBREVIATION': 'Abbreviation', 'TEAM_NAME': 'Team',
       'GAME_DATE': 'Date','WL': 'W/L','FG_PCT': 'FG%', 'FG3_PCT': 'FG3%','FT_PCT': 'FT%','PLUS_MINUS': '+/-'}).sort_values(by = 'Date')
print(stats.isnull().info())
stats
```

```python

stats.to_csv('clean_lakers_stats.csv')
```
